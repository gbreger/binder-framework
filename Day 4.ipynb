## Question 4

```python
# creates a groupby obj, based on NTA
nta_grp = df.groupby(['Neighborhood Tabulation Area (NTA) (2020)'])

# creates a filter that takes into account only the nta_grp that have at least 5 unique company names.
filt = nta_grp['Company Name'].unique().apply(len) >= 5
filt.value_counts()
```

```python
# returns the mean of total saving for every nta that meets the filt criteria (assigned above)
mean_total_savings = nta_grp['Total Savings'].mean().loc[filt]
mean_total_savings
```

```python
# returns the sum of jobs created for every nta that meets the filt criteria (assigned above)
sum_jobs = nta_grp['Job created'].sum().loc[filt]
sum_jobs
```

```python
output_dict = {
    'mean total savings': mean_total_savings,
    'sum of jobs created': sum_jobs
}
```

```python
output_df = pd.DataFrame(output_dict)
output_df.to_csv('desktop/data science/day_4_answers.csv')
```
